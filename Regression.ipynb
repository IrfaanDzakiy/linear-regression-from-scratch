Import module yang digunakan

```python
import Regression
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

Mengambil data dari csv dan memproses data non numeric

```python
df = Regression.processCSV("car_data.csv")
df
```

Split data 80% train data, 20% test data

```python
train, test = Regression.splitTestTrain(df, 0.2)
train
```

```python
test
```

Memisahkan attribut yang ingin di proses

```python
x_var, y_var = Regression.splitVar(train, ["Age", "Present_Price", "Kms_Driven/100"], ["Selling_Price"])
x_var[0:10]
```

```python
y_var[0:10]
```

```python
x_test, y_test = Regression.splitVar(test, ["Age", "Present_Price", "Kms_Driven/100"], ["Selling_Price"])
x_test[0:10]
```

```python
y_test[0:10]
```

Melakukan Training Data

```python
coef = Regression.calcCoeff(x_var, y_var)
coef
```

Melakukan Prediksi

```python
pred = Regression.predict(x_test, coef)
res = np.c_[y_test, pred]
res
```

Menunjukkan akurasi dari algoritma

```python
r2 = Regression.calc_rsquared(y_test, pred)
r2
```

Visualisasi Data

```python
Regression.drawReg(x_test, y_test, coef, 1, "Age", "Selling Price")
```

```python
Regression.drawReg(x_test, y_test, coef, 2, "Present_Price", "Selling Price")
```
